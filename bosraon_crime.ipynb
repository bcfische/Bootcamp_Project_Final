```python
def getLatitude(x):
    k = x.replace("(","")
    k = k.replace(")","")
    k = k.replace(" ","")
    k = k.split(',')
    k[0] = float(k[0])
#     print(k[0])
    return(k[0])
```

```python
def getLongitude(x):
    k = x.replace("(","")
    k = k.replace(")","")
    k = k.replace(" ","")
    k = k.split(',')
    k[1] = float(k[1])
#     print(k[1])
    return(k[1])
```

```python
getLatitude(data.iloc[1]["Location"])
```

```python
data["latitude"] = data['Location']
data["latitude"] =  data["latitude"].apply(lambda x: getLatitude(x))
```

```python
data["longitude"] = data['Location']
data["longitude"] =  data["longitude"].apply(lambda x: getLongitude(x))
```

```python
data = data.drop(columns=["Lat","Long","Location"])
data.head()
```
